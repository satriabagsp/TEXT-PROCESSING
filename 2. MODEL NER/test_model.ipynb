```python
import spacy
```

```python
# Load model NER masing-masing entitas
nlp_person = spacy.load('Person')
nlp_position = spacy.load('Position')
nlp_organization = spacy.load('Organization')
nlp_location = spacy.load('Location')
nlp_quote = spacy.load('Quote')
```

```python
# Contoh Teks
teks = 'Jakarta, CNBC Indonesia - Presiden Ukraina Volodymyr Zelensky meminta negara-negara penghasil energi untuk meningkatkan produksi minyak dan gas, demi menghindari "pemerasan" global dari Rusia. "Masa depan Eropa bergantung pada upaya Anda, saya meminta Anda untuk meningkatkan output energi untuk memastikan bahwa setiap orang Rusia memahami bahwa Anda tidak dapat menggunakan energi sebagai senjata," kata Zelensky dalam Forum Doha secara virtual, dikutip CNN Internasional, Sabtu (26/3/2022). "Qatar dapat memberikan kontribusi mereka pada stabilisasi Eropa," imbuhnya.'
teks
```

```python
# Mengaplikasikan setiap model NER ke variabel konten (isi berita)
doc_person = nlp_person(teks)
doc_position = nlp_position(teks)
doc_organization = nlp_organization(teks)
doc_location = nlp_location(teks)
doc_quote = nlp_quote(teks)

# Mengambil list entitas yang ditemukan oleh tiap model NER
person_temp = [(e.text) for e in doc_person.ents if e.label_ == 'person']
position_temp = [(e.text) for e in doc_position.ents if e.label_ == 'position']
organization_temp = [(e.text) for e in doc_organization.ents if e.label_ == 'organization']
location_temp = [(e.text) for e in doc_location.ents if e.label_ == 'location']
quote_temp = [(e.text) for e in doc_quote.ents if e.label_ == 'quote']
```
